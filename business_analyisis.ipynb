```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.prompts import PromptTemplate
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain.chains import create_retrieval_chain
from langchain.vectorstores import Chroma
```

```python
urls = ['https://www.ycombinator.com/companies/industry/generative-ai']# put your articles
```

```python
from langchain.document_loaders import UnstructuredURLLoader 
loaders = UnstructuredURLLoader(urls=urls)# now we are loading the website. 
data = loaders.load()
data
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-pro",
                             temperature=0.7)
embeddings = GoogleGenerativeAIEmbeddings(model='models/embedding-001')
```

```python
vecotrdb= Chroma.from_documents(data,embeddings)
```

```python
retriever = vecotrdb.as_retriever(search_kwargs={"k": 1})
```

```python
template = """
you are business advisor. you know the data of companies which are working on AI. use your knowledge to help.
context: {context}
input: {input}
answer:
"""
prompt = PromptTemplate.from_template(template)
combine_docs_chain = create_stuff_documents_chain(llm, prompt)
retrieval_chain = create_retrieval_chain(retriever, combine_docs_chain)
```

```python
response = retrieval_chain.invoke({'input':'suggest some business idea of startup ai'})
```
